Exploring the relationship between the target variable and other predictors.

### Numerical predictors

```python
train_data.select_dtypes(include=['int64', 'float64']).columns
```

Recode as categorical variables :`numPhotos`,`rooms`, `bathrooms`.

```python
train_data['rooms'] = train_data['rooms'].astype(str)
train_data['bathrooms'] = train_data['bathrooms'].astype(str)
```

Analysis of the target variable with a predictive variable

```python
def bivariate_numerical(df):
    target_variable = 'Price_Category'
    num_vars = train_data.select_dtypes(include=['int64', 'float64']).columns

    for num_var in num_vars:
        # Pairplot with hue
        sns.pairplot(df, hue=target_variable, vars=[num_var])
        plt.show()

        # Box plot categorized by the categorical variable
        sns.boxplot(x=target_variable, y=num_var, data=df)
        plt.show()

    # Adjust layout to ensure proper alignment of plots
    plt.subplots_adjust(wspace=0.5, hspace=0.5)
```

```python
bivariate_numerical(train_data)
```

### Categorical predictors

```python
def bivariate_categorical(df):
    target_variable = 'Price_Category'
    cat_vars = train_data.select_dtypes(include=['object']).drop(columns=["Price_Category"]).columns.tolist()

    # Define the layout of the subplots
    num_rows = 4
    num_cols = 4

    # Create subplots
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 15))

    # Flatten axes for easier indexing
    axes = axes.flatten()

    # Iterate over categorical variables and create plots
    for i, cat_var in enumerate(cat_vars):
        # Count plot for target variable grouped by categorical variable
        sns.countplot(x=cat_var, hue=target_variable, data=df, ax=axes[i])
        axes[i].set_title(f'{cat_var} vs {target_variable}')
        axes[i].set_xlabel(cat_var)
        axes[i].set_ylabel('Count')
        axes[i].tick_params(axis='x', rotation=45)
        axes[i].legend(title=target_variable)

    # Remove extra subplots
    for j in range(len(cat_vars), num_rows * num_cols):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()
```

```python
bivariate_categorical(train_data)
```

```python
# One-hot encode categorical variables
train_data_encoded = pd.get_dummies(train_data)
```

```python

# Calculate correlation matrix
correlation_matrix = train_data_encoded.corr()

# Filter correlations greater than 80%
high_correlations = correlation_matrix[abs(correlation_matrix) > 0.5]

# Plot heatmap
plt.figure(figsize=(30, 30))
sns.heatmap(high_correlations, annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```

```python
# List to store highly correlated pairs
highly_correlated_pairs = []

# Variables you are interested in
target_variables = ['Price_Category_High', 'Price_Category_Medium-High', 'Price_Category_Medium-Low','Price_Category_Low']  # Add other variables as needed

# Iterate over each column in the filtered correlation matrix
for col in high_correlations.columns:
    # Check if the current column is one of the target variables
    if col in target_variables:
        # Get indices of variables with correlation greater than 0.8
        correlated_vars = high_correlations.index[high_correlations[col].abs() > 0.5].tolist()
        # Remove the current variable from the list
        correlated_vars.remove(col)
        # Add pairs to the list
        for correlated_var in correlated_vars:
            # Check if the correlated variable is also in the target variables
            if correlated_var in target_variables:
                highly_correlated_pairs.append((col, correlated_var))

# Print the list of highly correlated variable pairs
print("Highly correlated variable pairs:")
for pair in highly_correlated_pairs:
    print(pair)
```

```python
def update_outliers_count(dataframe, column_name):
    df = dataframe.copy()
    sm = df[column_name].describe()
    iqr = sm['75%'] - sm['25%']
    # Severe Outliers
    severe_ub = sm['75%'] + 3 * iqr
    severe_lb = sm['25%'] - 3 * iqr
    severe_outliers_id = df[(df[column_name] > severe_ub) | (df[column_name] < severe_lb)].index
    df.loc[severe_outliers_id, 'univ_outl_count'] += 1
    return df
```
