```python
import pandas as pd
from google.colab import drive

import io
```

```python
# URL of the dataset
url = 'https://opendata-ajuntament.barcelona.cat/resources/bcn/EstadisticaPadro/pad/2023/2023_pad_mdb_nacionalitat-contintent_edat-q_sexe.csv'

# Read the data into a DataFrame
df_poblacio = pd.read_csv(url)
df_poblacio.drop(columns=['Codi_Districte', 'Codi_Barri', 'Data_Referencia'], inplace=True)

# URL of the dataset
url = 'https://opendata-ajuntament.barcelona.cat/resources/bcn/EstadisticaPadro/dimensions/pad_dimensions.csv'

# Read the data into a DataFrame
df_dimensions = pd.read_csv(url)
```

```python
df_dimensions
```

```python
# Decode df_poblacio with df_dimensions
for dimension in ['NACIONALITAT_CONTINENT', 'EDAT_Q', 'SEXE']:
    # Filter df_dimensions for the current dimension
    df_dim_filter = df_dimensions[df_dimensions['Desc_Dimensio'] == dimension]

    # Merge with df_poblacio on the code column
    df_poblacio = pd.merge(df_poblacio, df_dim_filter, left_on=dimension, right_on='Codi_Valor', how='left')

    # Replace the code with the description
    df_poblacio[dimension] = df_poblacio['Desc_Valor_EN']

    # Drop extra columns from the merge
    df_poblacio.drop(['Codi_Dimensio', 'Desc_Dimensio', 'Codi_Valor', 'Desc_Valor_CA', 'Desc_Valor_ES', 'Desc_Valor_EN'], axis=1, inplace=True)

print(df_poblacio)
```

```python
df_poblacio.to_csv('/content/drive/MyDrive/ML_Project/poblacio_barcelona.csv')
```
