```python
import base64
import requests
import json
from datetime import date
import numpy as np
from ast import literal_eval
```

```python
def get_access_token():
  api_key = 'KEY'
  secret = 'SECRET'
  message = api_key + ":" + secret

  auth = "Basic " + base64.b64encode(message.encode("ascii")).decode("ascii")

  headers_dic = {"Authorization" : auth,
                  "Content-Type" : "application/x-www-form-urlencoded;charset=UTF-8"}

  params_dic = {"grant_type" : "client_credentials",
                "scope" : "read"}

  r = requests.post("https://api.idealista.com/oauth/token",
                    headers = headers_dic,
                    params = params_dic)

  token = json.loads(r.text)['access_token']

  return token
```

```python
base_url = 'https://api.idealista.com/3.5/'
country = 'es' # Search country (values: es, it, pt)
language = 'en' # Search language (values: es, it, pt, en, ca)
max_items = '50' # Max items per call, the maximum set by idealista is 50
operation = 'sale' # Kind of operation (sale, rent)
property_type = 'homes' # (values: homes, offices, premises, garages, bedrooms)
order = ''
center = '41.373026,2.143878' # geographic coordinates (WGS84) (latitude, longitude)
distance = '1200'
sort = 'asc'
# maxprice = '750'
```

```python
def define_search_url():
    url = (base_url+country+
           '/search?operation='+operation+
           '&maxItems='+max_items+
        #    '&order='+order+
           '&center='+center+
           '&distance='+distance+
           '&propertyType='+property_type+
           '&sort='+sort+
           '&numPage=%s'+
        #    '&maxPrice='+maxprice+
           '&language='+language)
    return url
```

```python
url = define_search_url()
url
```

```python
def search_api(url):
    token = get_access_token()
    print(token)
    headers = {'Content-Type': 'Content-Type: multipart/form-data;', 'Authorization' : 'Bearer ' + token}
    content = requests.post(url, headers = headers)
    result = json.loads(content.text)

    return result
```

```python
# Search first page
page = 1
first_search_url = url %(page)
```

```python
results = search_api(first_search_url)
```

```python
def results_to_df(results):
    df = pd.DataFrame.from_dict(results['elementList'])
    return df
```

```python
def concat_df(df, df_tot):
    df_tot = pd.concat([df_tot,df])
    return df_tot
```

```python
df_tot = pd.DataFrame()
df_tot = concat_df(df, df_tot)
df_tot
```

```python
for i in range(2, total_pages + 1):
  url_page = url %(i)
  results = search_api(url_page)
  filename = f"{today}_{location}_{i}.json"
  path = f"{base_path_JSON}/{filename}"
  with open(path, 'w') as file:
    json.dump(results, file)

  df = results_to_df(results)
  df_tot = concat_df(df, df_tot)
```
